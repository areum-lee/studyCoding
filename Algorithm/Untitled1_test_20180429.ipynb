```python
import numpy as np

def memcpy(v, dst, src, size):    
    dst = int(dst)
    src = int(src)
    size = int(size)
    
    len_v = len(v)
    
    if size <= 0:
        print_result (v)
    elif len_v < src + size:
        print ('may throw RuntimeException or do nothing')
    
    else:
        temp = v[src:src+size]
        v[dst:dst+size] = temp
        print_result (v)
        
def print_result (v):
    len_v = len(v)
    temp_str = ""
    for i in range (len_v-1):
        temp_str += '%d' % (v[i]) + " "
    
    temp_str += '%d' % (v[len_v-1])
                        
    print (temp_str)
        
def get_int_from_str (str_input):
    temp_str_split = str_input.split(' ')
    temp_int = np.zeros ( len(temp_str_split) )
    for i in range (len(temp_str_split)):
        temp_int[i] = int (temp_str_split[i])
    return temp_int

n_input = int(input())

temp = input().split()
byte_array = np.zeros(len(temp))



for i in range (n_input):
    byte_array[i] = int(temp[i])

v = byte_array

str_input2 = input()
str_input2 = get_int_from_str (str_input2)

memcpy(v, str_input2[0], str_input2[1], str_input2[2])
```

```python
import sys

stack={}
key=[]
finish="1"

def evict():
    a=key.pop()
    del stack[a]  
def add(keykey, value):
    stack[keykey]=value
    key.append(keykey)  
def get(keykey):
    if keykey in key:
        print (stack[keykey])
    else:
        print("-1")  
def remove(keykey):
    if len(stack)==0:
        print("-1")
    else:
        print (stack[keykey])
    del stack[keykey]  
while finish != 0:
    order = input().split()

    if order[0]=="evict":
        evict()
    elif order[0]=="add":
        assert int(order[1])>0 and int(order[1])<=pow(2,30)
        assert int(order[2])>0 and int(order[2])<=pow(2,30)
        add(order[1],order[2])
    elif order[0]=="get":
        assert int(order[1])>0 and int(order[1])<=pow(2,30)
        get(order[1])    
    elif order[0]=="remove":
        remove(order[1])
    elif order[0]=="exit":
        finish = 0
```
